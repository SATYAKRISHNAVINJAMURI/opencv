## OpenCV BGR color palette

```python
import cv2 as cv;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
imgpath = "/home/satya/opencv/misc/girl.tiff";
img = plt.imread(imgpath);
r,g,b= cv.split(img);
plt.subplot(1,3,1);
plt.imshow(r,cmap="Reds");
plt.subplot(1,3,2);
plt.imshow(b,cmap="Blues");
plt.subplot(1,3,3);
plt.imshow(g,cmap="gray");
plt.show();
cv.bitwise_and()
```

```python
img = np.zeros((512,512,3),np.uint8);
```

```python
cap = cv.VideoCapture(0);
windowName = "My Video";
cv.namedWindow(windowName);
if cap.isOpened():
    ret,img = cap.read();
    print("width" + str(cap.get(3)));
    print("Height" + str(cap.get(4)));
    cap.set(3,640);
    cap.set(4,480);
    print("width" + str(cap.get(3)));
    print("Height" + str(cap.get(4)));
else:
    print("ERROR");

while ret:
        ret,img = cap.read();
        cv.imshow(windowName,img);
        hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV);
        low = np.array([100,50,50]);
        high = np.array([180,255,255]);
        image_mask = cv.inRange(hsv,low,high);
        cv.imshow("Mask",image_mask);
        
        if(cv.waitKey(1) == 27):
            break;
#     img = cv.cvtColor(img,cv.COLOR_BGR2RGB);
#     plt.imshow(img);
#     plt.xticks([]);
#     plt.yticks([]);
#     plt.show;
cv.destroyAllWindows();
cap.release();
```
